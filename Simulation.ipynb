```python
import pandas as pd
from mmcqueues import MMCQueue, ARRIVAL_RATES, SERVICE_RATES


max_Wq = 0.5
output = {}
for l in ARRIVAL_RATES:
    output[l] = {}
    for mu in SERVICE_RATES:
        for c in range(1, 200):
            if c * mu - l < 0:
                continue
            else:
                q = MMCQueue(l, mu, c)
                if q.Wq < 0.5:
                    output[l][mu] = {'c': c, 'Wq': q.Wq}
                    break

pd.DataFrame(output).transpose()
```
